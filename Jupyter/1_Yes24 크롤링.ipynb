```python
def get_book_links(genre_num):
    page = range(1, 8)
    page_url = 'http://www.yes24.com/24/Category/Display/' + genre_num + '?PageNumber='
    page_url_list = []
    for p in page:
        page_url_list.append(page_url + str(p))
    #print(page_url_list)

    urls = []
    for url in page_url_list:
        rand_value = np.random.randint(1, 3)
        time.sleep(rand_value)

        req = requests.get(page_url)
        soup = BeautifulSoup(req.text, 'html.parser')
# Yes24 카테고리별 베스트셀러 크롤링
- 국내도서 하위메뉴 한정. 베스트셀러
    - 소설/시/희곡
    - 에세이
    - 인문
    - 역사
    - 예술
    - 종교
    - 사회 정치
    - 자연과학
    - 경제 경영
    - 자기계발
- 각 카테고리당 (15 * 40 = 600권씩 수집)
    - 총 10 * (15 * 40) = 6000개의 데이터
        #url = soup.find('div', id='category_layout').find_all('li')
        url = soup.select('#category_layout div.goods_name > a[target="_blank"]')

        for i in url:
            urls.append(i['href'])

    return(urls)
```

```python
stopwords = {'한', '그는', '그녀는', '이', '통해', '그가', '무엇', '어디', '언제', '그녀가', '그들은', '그', '등', '이를', '세', '두', '더', '통해', '권', '명', '다', '때', '전', '있다', '것이다', '한다', '수', '있고', '있다', '많이',  '좀', '원', '많이', '나를', '그를', '그녀를', '나는', '내가', '그들을', '당신', '있는'}
okt = Okt()

def crawl_book(urls, genre, literature):
    book_for_genre = []
    for url in urls:
        rand_value = np.random.randint(1, 3)
        time.sleep(rand_value)
        basic_url = 'http://www.yes24.com'
        req = requests.get(basic_url + url)
        soup = BeautifulSoup(req.text, 'html.parser')

       # 제목
        try:
            title = soup.select_one('#yDetailTopWrap > div.topColRgt > div.gd_infoTop > div > h2').text
        except:
            title = ''

        # 저자
        author = ''
        try:
            author_list = soup.select("#yDetailTopWrap > div.topColRgt > div.gd_infoTop > span.gd_pubArea > span.gd_auth > a[href*='author_yn=Y']")
            for a in author_list:
                author += a.text + ' '
        except:
            author = ''
            
        # 별점
        try:
            rating = soup.select_one('#spanGdRating > a > em').text
        except:
            rating = ''
            
        # 카테고리
        category = ''
        category_list = []
        try: 
            category_total = soup.select('#infoset_goodsCate > div.infoSetCont_wrap > dl.yesAlertDl > dd > ul > li > a[href*="Category/Display"]')
            #print(category_total)
            for c in category_total:
                if c.text == '소설/시/희곡':
                    category_list.append(literature)
                    continue
                category_list += c.text.split('/')
            #print(set(category_list))
            category = ' '.join(set(category_list))
        except:
            category = ''
            
        # 태그
        tag = ''
        try:
            tag_list = soup.select('#infoset_goodsCate > div.infoSetCont_wrap > dl > dd > span > a[href*="keyword"]')
            for t in tag_list:
                tag += re.sub(r'[@\\*=()/~#&\+á?\xc3\xa1\-\|\:\;\!\-\,\_\~\$\'\"…‘’“”《》]', ' ', t.text)
        except:
            tag = ''
            
        # 책 소개 (키워드)
        try: 
            overview = soup.select_one('#infoset_introduce > div.infoSetCont_wrap > div.infoWrap_txt > div').text.strip().replace('\n', ' ').replace('\r', ' ')
        except:
            keyword = ''
            book_for_genre.append(
                {
                    'title': title,
                    'author': author,
                    'rating': rating,
                    'category': category,
                    'tag': tag,
                    'keyword': keyword,
                }
            )
            continue
        overview = re.sub(r'[@\\*=()/~#&\+á?\xc3\xa1\-\|\:\;\!\-\,\_\~\$\'\"…‘’“”《》]', '', overview)
        overview = re.sub(r'\s\s+', ' ', overview)
        overview_list = overview.split('. ')
        keyword = ''
        for overview in overview_list:
            overview = okt.pos(overview, stem=True)
            word = []
            for o in overview:
                if not o[1] == 'Noun':
                    continue
                if len(o[0]) == 1:
                    continue
                word.append(o[0])
            word = ' '.join(word)
            word += '. '
            keyword += word
        texts = keyword.split('. ')
        try:
            keywords = summarize_with_keywords(texts, min_count=3, max_length=10, beta=0.85, stopwords=stopwords, max_iter=10, verbose=True)
        except ValueError:
            #print('key가 없습니다.')
            keyword = ''
            book_for_genre.append(
                {
                    'title': title,
                    'author': author,
                    'rating': rating,
                    'category': category,
                    'tag': tag,
                    'keyword': keyword,
                }
            )
            continue
        keyword = ''
        for word, r in sorted(keywords.items(), key=lambda x:x[1], reverse=True)[:10]:
            #print(word + ': ' + str(r))
            keyword += word + ' '
        
        
        book_for_genre.append(
            {
                'title': title,
                'author': author,
                'rating': rating,
                'category': category,
                'tag': tag,
                'keyword': keyword,
            }
        )
    
    df = pd.DataFrame(book_for_genre)
    df.to_csv("./crawled_data/csv/book_info_"+genre+".csv")
    df.to_pickle("./crawled_data/pkl/book_info_"+genre+".pkl")
    return(df)
```

```python
def get_book_links(genre_num):
    page = range(1, 16)
    page_url = 'http://www.yes24.com/24/Category/More/' + genre_num + '?ElemNo=104&ElemSeq=1&FetchSize=40&PageNumber='
    page_url_list = []
    for p in page:
        page_url_list.append(page_url + str(p))
    #print(page_url_list)

    urls = []
    for url in page_url_list:
        rand_value = np.random.randint(1, 3)
        time.sleep(rand_value)

        req = requests.get(url)
        soup = BeautifulSoup(req.text, 'html.parser')

        #url = soup.find('div', id='category_layout').find_all('li')
        url_books = soup.select('#category_layout div.goods_name > a[target="_blank"]')

        for i in url_books:
            urls.append(i['href'])

    return(urls)
```

```python
stopwords = {'한', '그는', '그녀는', '이', '통해', '그가', '무엇', '어디', '언제', '그녀가', '그들은', '그', '등', '이를', '세', '두', '더', '통해', '권', '명', '다', '때', '전', '있다', '것이다', '한다', '수', '있고', '있다', '많이',  '좀', '원', '많이', '나를', '그를', '그녀를', '나는', '내가', '그들을', '당신', '있는'}
okt = Okt()

def crawl_book(urls, genre, literature):
    book_for_genre = []
    for url in urls:
        rand_value = np.random.randint(1, 3)
        time.sleep(rand_value)
        basic_url = 'http://www.yes24.com'
        req = requests.get(basic_url + url)
        soup = BeautifulSoup(req.text, 'html.parser')

       # 제목
        try:
            title = soup.select_one('#yDetailTopWrap > div.topColRgt > div.gd_infoTop > div > h2').text
        except:
            title = ''

        # 저자
        author = ''
        try:
            author_list = soup.select("#yDetailTopWrap > div.topColRgt > div.gd_infoTop > span.gd_pubArea > span.gd_auth > a[href*='author_yn=Y']")
            for a in author_list:
                author += a.text + ' '
        except:
            author = ''
            
        # 별점
        try:
            rating = soup.select_one('#spanGdRating > a > em').text
        except:
            rating = ''
            
        # 카테고리
        category = ''
        category_list = []
        try: 
            category_total = soup.select('#infoset_goodsCate > div.infoSetCont_wrap > dl.yesAlertDl > dd > ul > li > a[href*="Category/Display"]')
            #print(category_total)
            for c in category_total:
                if c.text == '소설/시/희곡':
                    category_list.append(literature)
                    continue
                category_list += c.text.split('/')
            #print(set(category_list))
            category = ' '.join(set(category_list))
        except:
            category = ''
            
        # 태그
        tag = ''
        try:
            tag_list = soup.select('#infoset_goodsCate > div.infoSetCont_wrap > dl > dd > span > a[href*="keyword"]')
            for t in tag_list:
                tag += re.sub(r'[@\\*=()/~#&\+á?\xc3\xa1\-\|\:\;\!\-\,\_\~\$\'\"…‘’“”《》]', ' ', t.text)
        except:
            tag = ''
            
        # 책 소개 (키워드)
        try: 
            overview = soup.select_one('#infoset_introduce > div.infoSetCont_wrap > div.infoWrap_txt > div').text.strip().replace('\n', ' ').replace('\r', ' ')
        except:
            keyword = ''
            book_for_genre.append(
                {
                    'title': title,
                    'author': author,
                    'rating': rating,
                    'category': category,
                    'tag': tag,
                    'keyword': keyword,
                }
            )
            continue
        overview = re.sub(r'[@\\*=()/~#&\+á?\xc3\xa1\-\|\:\;\!\-\,\_\~\$\'\"…‘’“”《》]', '', overview)
        overview = re.sub(r'\s\s+', ' ', overview)
        overview_list = overview.split('. ')
        keyword = ''
        for overview in overview_list:
            overview = okt.pos(overview, stem=True)
            word = []
            for o in overview:
                if not o[1] == 'Noun':
                    continue
                if len(o[0]) == 1:
                    continue
                word.append(o[0])
            word = ' '.join(word)
            word += '. '
            keyword += word
        texts = keyword.split('. ')
        try:
            keywords = summarize_with_keywords(texts, min_count=3, max_length=10, beta=0.85, stopwords=stopwords, max_iter=10, verbose=True)
        except ValueError:
            #print('key가 없습니다.')
            keyword = ''
            book_for_genre.append(
                {
                    'title': title,
                    'author': author,
                    'rating': rating,
                    'category': category,
                    'tag': tag,
                    'keyword': keyword,
                }
            )
            continue
        keyword = ''
        for word, r in sorted(keywords.items(), key=lambda x:x[1], reverse=True)[:10]:
            #print(word + ': ' + str(r))
            keyword += word + ' '
        
        
        book_for_genre.append(
            {
                'title': title,
                'author': author,
                'rating': rating,
                'category': category,
                'tag': tag,
                'keyword': keyword,
            }
        )
    
    df = pd.DataFrame(book_for_genre)
    df.to_csv("./crawled_data/csv/book_info_"+genre+".csv")
    df.to_pickle("./crawled_data/pkl/book_info_"+genre+".pkl")
    return(df)
```
